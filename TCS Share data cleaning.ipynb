```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as date

import matplotlib 
import matplotlib.pyplot as plt
```

```python
# Import the data from the excel with the help  pandas function read_csv()

df = pd.read_csv(r"C:\Users\HP\Desktop\New folder\TCS Share.csv")
```

```python
# With the help of head() function lets check the first four rows of imported data

df.head()
```

```python
# Findiing is there any messing value in the data if yes, how many percentage of missing data per coloumn

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print("{}---{}%".format(col, pct_missing))
    
# As we can see in the result that the percentage of missing value in the data is zero
```

```python
# Let's see if the data type of the columns is correct or not

df.dtypes
```

```python
# Changing the typ of the nessesay column

df['Date'] = pd.to_datetime(df['Date'], format="mixed")
df['Change(%)'] = df['Change(%)'].str.replace('%', '')
df['Change(%)'] = df['Change(%)'].astype(float)
```

```python
# Now we dont wnat any duplicate value to be present in the data, so we drop all the duplicate values

df.drop_duplicates()
```

```python
df
```

```python
# Creating a new column called DayName which have the dayname of the particular date

df['DayName'] = df['Date'].dt.day_name()
```

```python
# Creating a new column called Year from the date column

df['Year'] = df['Date'].dt.year
```
